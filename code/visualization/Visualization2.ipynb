```python
import numpy as np
import pandas as pd
import glob
```

```python
records = pd.read_csv(r'C:\Users\User\Desktop\Stanley\Testing\Data\SHC1067Rrecords.txt', sep = ",", header = None)
labels = pd.read_csv(r'C:\Users\User\Desktop\Stanley\Testing\Data\SHC1067Rlabels.txt', sep = ",", header = None, usecols=[0])
pred = pd.read_csv(r'C:\Users\User\Desktop\Stanley\Testing\Data\SHC1067Rpredictions.txt', sep = ",", header = None)
prob = pd.read_csv(r'C:\Users\User\Desktop\Stanley\Testing\Data\SHC1067Rprobabilities1.txt', sep = ",", header = None)
```

```python
records.columns = ["Mean", "Stdev", "Over60", "O60U120", "Over120", "Skewness", "Kurtosis", "Entropy", "Q1", "Q2", "Q3", "Q4", "Q5", "Q6", "Q7", "Q8", "Q9"]
pred.columns = ["Pred"]
prob.columns = ["Prob"]
```

```python
## Appending labels and predictions together for visualization ##
df = pd.concat([pd.DataFrame(labels.values), pd.DataFrame(pred.values)], axis=1)
df.columns = ["Labels", "Prediction"]
df
```

```python
## % of correct predictions ##
count = 0.0
for i in range(len(df)):
    if df.iloc[i, 0] == df.iloc[i,1]:
        count = count + 1

print(count / len(df))
```

```python
def correct(df1, df2):
    count = 0.0
    size = len(df2)
    if (len(df1) != size):
        return -1
    for i in range(size):
        if int(df1.iloc[i, 0]) == int(df2.iloc[i, 0]):
            count = count + 1

    return count / size

val = correct(labels, pred)
val
```

```python
## Sorting based on prob values in descending order ##
prob = prob.sort_values('Prob', ascending=False)
prob
```
